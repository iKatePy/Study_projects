Шаг 5. Проверка гипотез
Нулевая гипотеза: Нет значимой разницы в среднем доходе между клиентами с разным уровнем образования.

Альтернативная гипотеза: Существует значимая разница в среднем доходе между клиентами с разным уровнем образования.

Визуализация данных

```python
# Boxplot для Income по Education
sns.boxplot(x='Education', y='Income', data=data)
plt.show()
```

Выбор статистического критерия


Используем ANOVA для сравнения средних доходов между группами. ANOVA позволяет одновременно проверить гипотезу о равенстве средних доходов между всеми группами, тогда как t-тест требует множественных сравнений, что увеличивает риск ложноположительных результатов

```python
from scipy.stats import f_oneway

# Разделение данных по группам
groups = [data[data['Education'] == edu]['Income'] for edu in data['Education'].unique()]

# ANOVA
f_stat, p_value = f_oneway(*groups)
print(f"F-статистика: {f_stat}, p-значение: {p_value}")
```

Вывод
Если p-значение < 0.05, отвергаем нулевую гипотезу и делаем вывод, что существует значимая разница в среднем доходе между группами.

Шаг 6. Построение модели кластеризации

Метод локтя для определения числа кластеров

```python
from sklearn.cluster import KMeans # Используем метод k - средних
import numpy as np

# Выбор признаков для кластеризации, финансовые показатели, которые могут влиять на поведение клиентов
# Кластеризация по этим признакам поможет выявить типы клиентов с похожими покупательскими привычками и доходами
X = data[['Income', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts']]

# Метод локтя
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_) # сумма квадратов расстояний точек до их центров кластеров — сохраняется в список

# Визуализация метода локтя
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

В точке 4 inertia перестает существенно уменьшаться

Кластеризация K-means

```python
# Выбор числа кластеров
kmeans = KMeans(n_clusters=4, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)

# Удаление строковых столбцов перед анализом кластеров
numeric_data = data.drop(columns=['Education', 'Marital_Status', 'Dt_Customer'])

# Анализ кластеров
cluster_analysis = numeric_data.groupby('Cluster').mean()
print(cluster_analysis)
```
